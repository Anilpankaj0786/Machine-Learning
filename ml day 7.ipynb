```python
numerical_imputer = SimpleImputer(strategy='mean')
categorical_imputer = SimpleImputer(strategy='most_frequent')
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns
```

```python
df[numerical_columns] = numerical_imputer.fit_transform(df[numerical_columns])
df[categorical_columns] = categorical_imputer.fit_transform(df[categorical_columns])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_label= df.copy()
```

```python
lb=LabelEncoder()
```

```python
df_label['gender'] = lb.fit_transform(df_label['gender'])
df_label['enrolled_university'] = lb.fit_transform(df_label['enrolled_university'])
df_label['education_level']= lb.fit_transform(df_label['education_level'])
df_label['company_type'] = lb.fit_transform(df_label['company_type'])
df_label['major_discipline'] = lb.fit_transform(df_label['major_discipline'])
df_label['relevent_experience'] = lb.fit_transform(df_label['relevent_experience'])
df_label['city'] = lb.fit_transform(df_label['city'])
df_label['last_new_job'] = lb.fit_transform(df_label['last_new_job'])
```
