```python
# The formula for Laplace smoothing in the context of Naive Bayes is:
# P(w|c) = (count(w,c) + alpha) / (count(c) + alpha * num_features).
# P(w|c): Probability of word w given class c.
# count(w,c): Number of times word w appears in class c.
# alpha: The smoothing parameter (typically set to 1).
# num_features: The total number of unique features (words) in the vocabulary.
```

```python
from sklearn. feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
# dataset
texts = [
"buy cheap meds now",
"cheap pills online",
"get discount now",
"meeting at office",
"project discussion",
"schedule a meeting"
]
labels = ["Spam", "Spam", "Spam", "Not Spam", "Not Spam", "Not Spam"]
# Create a model pipeline
model = make_pipeline(CountVectorizer(), MultinomialNB(alpha=1.0)) # Laplace smoothness
# Train the model
model.fit(texts, labels)

# Test messages
test_messages = [
"cheap meds now",
"schedule a meeting",
"project schedule",
"discount online pills",
"meeting now"
]
# Make predictions
predicted = model.predict(test_messages)
probs = model. predict_proba(test_messages)
# Display results
for msg, label, prob in zip(test_messages, predicted, probs):
    print(f"Message: '{msg}'")
    print(f"Predicted Class: {label}")
    print(f"Probabilities: Spam={prob[model.classes_.tolist().index('Spam')]:.2f}, Not Spam={prob[model.classes_.tolist().index('Not Spam')]:.2f}\n")
```
